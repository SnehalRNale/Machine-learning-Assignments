```python
x=df["YEAR"]
y=df["ANNUAL"]
```

```python
plt.plot(x,y,'o')
```

```python
sns.scatterplot(x=x,y=y,data=df)
```

```python
x.shape
```

```python
x=x.values
x=x.reshape(117,1)
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
print(f"x training Dataset: {x_train.shape}")
print(f"y Training Dataset: {y_train.shape}")
print(f"x Testing Dataset: {x_test.shape}")
print(f"y Testing Dataset: {y_test.shape}")
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred.shape
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_test,y_pred,color="Black",linewidth=3)
plt.title("Temperature V/s Year")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

```python
sns.regplot(data=df,x=x_train,y=y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MSE :{mean_squared_error(y_test,y_pred)}")
print(f"MAE :{mean_absolute_error(y_test,y_pred)}")
print(f"R2 :{r2_score(y_test,y_pred)}")
```
